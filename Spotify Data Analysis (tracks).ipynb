### We will analyze the Spotify music platform and draw valuable insight

**We will use this dataset gotten from Kaggle.com for this analysis - https://www.kaggle.com/lehaknarnauli/spotify-datasets?select=artists.csv**

```python
#Import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#read the file into a pandas dataframe
tracks = pd.read_csv('tracks.csv')
```

```python
#first, let's get details about the data
tracks.info()
```

```python
#Let's check for null/missing values 
pd.isnull(tracks).sum()
```

**We see above that there is no missing values in our data except the 'name' column**

### Let's get some insights now

**First we check the 10 least popular songs in the dataset**

```python
tracks.sort_values('popularity', ascending=True, inplace=False).head(10)
```

```python
#we can also check for the 10 most popular songs
most_popular = tracks.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

**Descriptive statistics for the numerical variables in the columns of our dataset**

```python
tracks.describe().transpose()
```

**Next thing we want to do is to change the index and set it as the release_date**

```python
tracks.set_index('release_date', inplace=True)

#we will also change the value of the new index to pandas datetime
tracks.index = pd.to_datetime(tracks.index)

#check the dataframe to see what we've done
tracks.head()
```

**If you observe, the duration of the songs are in milliseconds (duration_ms column), let's see if we can convert them into seconds. We will do that with lambda expresion**

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x: round(x/1000))

#we will drop the duration_ms column since we want the new duration to be there permanently
tracks.drop('duration_ms', axis=1, inplace=True)
```

```python
tracks.duration.head()
```

## Data Visualization

```python
#we will start our visualization with a correlation heatmap
#first, we drop 3 columns we will not be using 
track_corr = tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')

#set figure size
plt.figure(figsize=(14,6))

#we are using the seaborn heatmap to create the correlation map
heatmap = sns.heatmap(track_corr, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='magma', linewidths=1, linecolor='black')

#heatmap title
heatmap.set_title('Correlation Heatmap Between Varibles')

#set the labels on the x axis so they can show properly
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

**Now we create another column extracted from the release_date column**

```python
tracks['dates'] = tracks.index.get_level_values('release_date')

#set the new column to pandas datetime
tracks.dates = pd.to_datetime(tracks.dates)

#let's get the year out of that
years = tracks.dates.dt.year
```

**We will create a distribution plot to visualize the total number of songs per year**

```python
sns.displot(years, discrete=True, aspect=2, height=5, kind='hist', color='green').set(title='Songs per year')
```

**Now we will look at the duration of songs over the years via a barplot**

```python
#create a variable for the duration in seconds we have before
total_duration =tracks.duration

#set the figure parameters
fig_dims = (187)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years,y=total_duration, ax=ax, errwidth=False).set(title='Year Vs Duration')
plt.xticks(rotation=90)
```

**Finally, we will create a lineplot to analyse the average duration of the songs over the years**

```python
total_duration =tracks.duration
sns.set_style(style='whitegrid')
fig_dims = (13,6)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.lineplot(x=years, y=total_duration, ax=ax).set(title='Year Vs Duration')
plt.xticks(rotation=60)
```
